```python
data_comment=pd.read_csv("comment_data.csv")
```

```python
data_evaluation=pd.read_csv("evaluation.data.csv")
```

```python
data_all=pd.read_csv("overall.data.csv")
```

```python
data_all=data_all.drop(['Unnamed: 0'],axis=1)
```

```python
#### Adjusted rating data 
data_rating=pd.read_csv("adjusted_rating_data.csv")
```

```python
#### Combine adjusted rating into data set 
data_all_adjust=pd.merge(data_all,data_rating, on="Name")
```

```python
data_all_adjust.head(10)
```

```python
data_all_adjust.columns
```

```python
### Percentage of positive sentiment based on Doctors
positive_perc=data_all.groupby(["Name","Dominant_Topic"])["Positive Sentiment"].mean()
```

```python
positive_perc_data=pd.DataFrame(positive_perc)
```

```python
positive_perc_data.head(10)
```

```python
positive_perc_data=positive_perc_data.reset_index()
```

```python
#### Comment data 
data_comment=pd.read_csv("comment_data.csv")
```

```python
data_comment=data_comment.drop(['Unnamed: 0'],axis=1)
```

```python
data_comment
```

```python
data_new=pd.merge(positive_perc_data,data_comment,on=['Name','Dominant_Topic'])
```

```python
data_summary=pd.merge(data_new,data_all_adjust[["Name","Dominant_Topic","Speciality",'count', 'bayes']],on=['Name','Dominant_Topic'])
```

```python
data_summary=data_summary.drop_duplicates()
```

```python
data_summary=data_summary.drop(["index"],axis=1)
```

```python
data_summary.head(10)
```

```python
data_summary.columns=["Name","Dominant Topic"," Percentage of Positive Sentiment",
                      "Review Quality Index","Comment","Speciality", "Number of Review",
                      "Adjusted Rating Score"]
```
